```python
import zipfile
import os

# --- Define your file paths ---
model_zip_path = "/content/drive/MyDrive/Projects/DBMSProject/training_runs/run_7_MTID_Infra_4_classes2-20251028T025537Z-1-001.zip"
test_set_zip_path = "/content/drive/MyDrive/Projects/DBMSProject/Dataset Annotation/MTID/MTID_YOLO_Infra_Dataset/test-20251028T025801Z-1-001.zip"

# Define where to extract them
model_extract_path = "/content/drive/MyDrive/Projects/DBMSProject/training_runs/"
dataset_extract_path = "/content/drive/MyDrive/Projects/DBMSProject/Dataset Annotation/MTID/MTID_YOLO_Infra_Dataset/"

# --- Unzip the model ---
print(f"Extracting model from {model_zip_path}...")
try:
    with zipfile.ZipFile(model_zip_path, 'r') as zip_ref:
        zip_ref.extractall(model_extract_path)
    print("✅ Model unzipped successfully.")
except Exception as e:
    print(f"❌ Error unzipping model: {e}")

# --- Unzip the test set ---
print(f"\nExtracting test set from {test_set_zip_path}...")
try:
    with zipfile.ZipFile(test_set_zip_path, 'r') as zip_ref:
        zip_ref.extractall(dataset_extract_path)
    print("✅ Test set unzipped successfully.")
except Exception as e:
    print(f"❌ Error unzipping test set: {e}")
```
